```python
#Write a custom iterator class that prints numbers from 1 to 5
```

```python
class Numbers:
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <= 5:
            value = self.num
            self.num += 1
            return value
        else:
            raise StopIteration

obj = Numbers()

for i in obj:
    print(i)
```

```python
#Write an iterator that returns only even numbers up to a limit.
```

```python
class EvenNumbers:
    def __init__(self, limit):
        self.limit = limit
        self.num = 2

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= self.limit:
            value = self.num
            self.num += 2
            return value
        else:
            raise StopIteration

obj = EvenNumbers(10)

for i in obj:
    print(i)
```
