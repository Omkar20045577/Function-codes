```python
#square a number
```

```python
square = lambda x: x * x

print(square(5))
```

```python
#even or odd
```

```python
check = lambda x:"Even" if x % 2 == 0 else"odd"

print(check(10))
print(check(7))
```

```python
#list of tuples based on the second element
```

```python
students=[("omkar",85),("sai",92),("vardhan",78)]
sorted_list = sorted(students,key = lambda x : x[1])
print(sorted_list)
```

```python
#Maximum of two numbers
```

```python
Maximum = lambda a,b:a if a>b else b
print(Maximum(10,20))
```

```python
#sorted words by their length
```

```python
words = ["Apple","kiwi","banana","grape"]
sorted_words = sorted(words,key = lambda x : len(x))
print(sorted_words)
```
