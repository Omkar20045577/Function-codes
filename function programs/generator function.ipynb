```python
#Write a generator function to generate numbers from 1 to 10.
```

```python
def generate_numbers():
    for i in range(1, 11):
        yield i

for num in generate_numbers():
    print(num)
```

```python
#Create a generator for even numbers between 1 and 20.
```

```python
def even_numbers():
    for i in range(2, 21, 2):
        yield i

for num in even_numbers():
    print(num)
```

```python
#Write a Fibonacci sequence generator.
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num)
```

```python
#Create a generator expression to square numbers.
```

```python
squares = (x**2 for x in range(1, 11))

for sq in squares:
    print(sq)
```

```python
#Write a generator that yields vowels from a string.
```

```python
def vowels_in_string(s):
    vowels = "aeiouAEIOU"
    for ch in s:
        if ch in vowels:
            yield ch

text = "Kittu loves Python Generators"
for v in vowels_in_string(text):
    print(v)
```
