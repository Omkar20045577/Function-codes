```python
#Use filter() to get all even numbers from a list.
```

```python
numbers = [1,2,3,4,5,6,7,8]
even_numbers = list(filter(lambda x:x %2 == 0,numbers))
print(even_numbers)
```

```python
#Filter all names starting with the letter 'A'.
```

```python
names = ["Omkar", "Teja", "Arjun", "Ravi", "Anil"]

a_names = list(filter(lambda name: name.startswith('A'), names))

print(a_names)
```

```python
#Use filter() to remove negative numbers from a list.
```

```python
numbers = [-5,10,-15,-20,25,30]
positive_numbers = list(filter(lambda x: x >=0, numbers))
print(positive_numbers)
```

```python
#Filter words whose length is greater than 5.
```

```python
words = ["Omkar", "Vardhan", "ravi", "Bhaskar", "Teja"]
long_words = list(filter(lambda word: len(word)>5, words))
print(long_words)
```

```python
#From a list of ages, filter people eligible to vote (18+).
```

```python
ages = [15, 19, 18, 21, 17]

eligible_voters = list(filter(lambda age: age >= 18, ages))

print(eligible_voters)
```
