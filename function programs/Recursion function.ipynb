```python
#Write a recursive function to find factorial of a number.
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


print(factorial(5))  
```

```python
#Find Fibonacci numbers using recursion.
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Example: print first 10 Fibonacci number
for i in range(10):
    print(fibonacci(i))
```

```python
#Write a recursive function to calculate sum of digits of a number.
```

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)

# Example
print(sum_of_digits(1234))  
```

```python
#Reverse a string using recursion.
```

```python
def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

# Example
print(reverse_string("Omkar")) 
```

```python
#Find the power of a number using recursion.
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)

# Example
print(power(2, 5))  
```

```python
#Check whether a string is palindrome using recursion.
```

```python
def is_palindrome(s):
    if len(s) <= 1:   
        return True
    if s[0] != s[-1]: 
        return False
    return is_palindrome(s[1:-1])  

# Example
print(is_palindrome("madam"))  
print(is_palindrome("hello"))  
```

```python
#Find the greatest common divisor (GCD) using recursion.
```

```python
def gcd(a, b):
    if b == 0:   
        return a
    return gcd(b, a % b)  

# Example
print(gcd(48, 18))  
```

```python
#Solve Tower of Hanoi using recursion.
```

```python
def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n-1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n-1, auxiliary, target, source)

# Example
tower_of_hanoi(3, 'A', 'C', 'B')
```

```python
#Count the number of digits using recursion.
```

```python
def count_digits(n):
    if n == 0:   
        return 0
    return 1 + count_digits(n // 10)

# Example
print(count_digits(12345))  
```

```python
#Find the sum of elements in a list using recursion.
```

```python
def list_sum(lst):
    if not lst:   
        return 0
    return lst[0] + list_sum(lst[1:])  

# Example
print(list_sum([1, 2, 3, 4, 5])) 
```
