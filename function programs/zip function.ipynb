```python
#Combine two lists into a dictionary using zip().
```

```python
keys = ["name", "age", "city"]
values = ["Omkar", 21, "Guntur"]

result = dict(zip(keys, values))

print(result)
```

```python
#Zip two lists and print paired elements.
```

```python
names = ["Ravi", "Sai", "Vardhan"]
marks = [85, 90, 95]

for name, mark in zip(names, marks):
    print(name, "-", mark)
```

```python
#Unzip a list of tuples using zip().
```

```python
data = [("Omkar", 85), ("Sai", 90), ("Vardhan", 95)]
names, marks = zip(*data)

print("Names:", names)
print("Marks:", marks)
```

```python
#Use zip() to add corresponding elements from two lists.
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = [a + b for a, b in zip(list1, list2)]

print(result)
```

```python
#Iterate through two lists simultaneously using zip().
```

```python
students = ["Ravi", "Sai", "Teja"]
grades = ["A", "B", "A+"]

for student, grade in zip(students, grades):
    print(student, "got", grade)
```
