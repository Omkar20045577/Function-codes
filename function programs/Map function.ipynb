```python
#square all numbers in a list
```

```python
numbers = [1,2,3,4,5]
squares = list(map(lambda x:x**2, numbers))
print(squares)
```

```python
#strings into uppercase using map()
```

```python
words = ["omkar", "sai", "vardhan"]
upper_case_words = list(map(str.upper,words))
print(upper_case_words)
```

```python
#temperatures from Celsius to Fahrenheit
```

```python
celsius = [0, 20, 30, 40]

fahrenheit = list(map(lambda c: (9/5) * c + 32, celsius))

print(fahrenheit)
```

```python
#Add corresponding elements of two lists using map().
```

```python
list1 = [1,2,3]
list2 = [4,5,6]
result = list(map(lambda x,y:x+y,list1,list2))
print(result)
```

```python
#Convert a list of numeric strings into integers using map().
```

```python
numbers = ["1","2","3","4"]
integer_numbers = list(map(int,numbers))
print(integer_numbers)
```
